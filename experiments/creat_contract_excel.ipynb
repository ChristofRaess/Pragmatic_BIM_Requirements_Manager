```python
col_attributes = [
    'Sort',
    'PsetQsetName',
    'AttributeNameDE',
    'DescriptionDE',
    'Unit',
    'ValueDE',
    'IfcDataTypeName',
    'ElementNameDE',
    'ModelNameDE',
    'UsecaseNameDE',
    'PhaseDE',
    'HierarchyNameDE',
]
```

```python
attributes_df_filtered = attributes_df[col_attributes]
attributes_df_filtered.head()
```

```python
columns_to_explode = ['ElementNameDE', 'ModelNameDE', 'UsecaseNameDE']

# Split the comma-separated strings into lists
for column in columns_to_explode:
    attributes_df_filtered[column] = attributes_df_filtered[column].str.split(',')

# Explode the DataFrame on each column sequentially
for column in columns_to_explode:
    attributes_df_filtered = attributes_df_filtered.explode(column)

# Reset the index after exploding
attributes_df_filtered = attributes_df_filtered.reset_index(drop=True)

# Strip whitespace from the exploded values
for column in columns_to_explode:
    attributes_df_filtered[column] = attributes_df_filtered[column].str.strip()

#attributes_df_filtered = attributes_df_filtered.sort_values(by=['Sort'], 
#                                   ascending=[True])

attributes_df_filtered['Sort'] = pd.to_numeric(attributes_df_filtered['Sort'], errors='coerce')

# Sort, putting NaN values at the end
attributes_df_filtered = attributes_df_filtered.sort_values(by=['Sort'], ascending=[True], na_position='last')
attributes_df_filtered = attributes_df_filtered.rename(columns={'Sort': 'SortAttributes'})
# Convert to numeric, coercing errors to NaN

attributes_df_filtered.head()
```

```python
col_elements = [
    'Sort',
    'ElementNameDE',
    'ModelingGuidelineDE',
    'ElementDescriptionDE',
    'IfcEntity',
]


elements_df_filtered = elements_df[col_elements]

sort_by_colume(elements_df_filtered, 'Sort')
elements_df_filtered = elements_df_filtered.rename(columns={'Sort': 'SortElements'})
elements_df_filtered.head()
```

```python
elements_df_filtered['MergedColumn'] = np.where(
    elements_df_filtered['ModelingGuidelineDE'].notna(),
    elements_df_filtered['ElementDescriptionDE'].astype(str) + '\n\n' + elements_df_filtered['ModelingGuidelineDE'].astype(str),
    elements_df_filtered['ElementDescriptionDE']
)

# Remove the original columns if you don't need them anymore
elements_df_filtered = elements_df_filtered.drop(['ModelingGuidelineDE', 'ElementDescriptionDE'], axis=1)

# If you want to rename the new column
elements_df_filtered = elements_df_filtered.rename(columns={'MergedColumn': 'ElementDescriptionDE_new'})

elements_df_filtered.head()
```

```python
col_models= [
    'Sort',
    'ModelNameDE',
    'ModelDescribtionDE',
    'FileName',
]


models_df_filtered = models_df[col_models]
sort_by_colume(models_df_filtered, 'Sort')
models_df_filtered = models_df_filtered.rename(columns={'Sort': 'SortModels'})
models_df_filtered.head()
```

```python
merged_df = pd.merge(attributes_df_filtered, 
                     elements_df_filtered, 
                     on='ElementNameDE', 
                     how='left')


all_df = pd.merge(merged_df, 
                     models_df_filtered, 
                     on='ModelNameDE', 
                     how='left')


all_df.head()
```

```python
all_phases = set()
for phases in all_df['PhaseDE'].str.split(','):
    all_phases.update([phase.strip() for phase in phases if isinstance(phases, list)])

# Convert to a sorted list
all_phases = sorted(all_phases)

print("Unique phases:", all_phases)
```

```python
for phase in all_phases:
    all_df[f'Phase_{phase}'] = ''

# Fill the new columns with 'X' where appropriate
for index, row in all_df.iterrows():
    phases = row['PhaseDE'].split(',') if isinstance(row['PhaseDE'], str) else []
    for phase in phases:
        phase = phase.strip()
        if phase in all_phases:
            all_df.at[index, f'Phase_{phase}'] = 'X'

all_df.head()
```

```python
# Check if the necessary sort columns are available
columns_to_check = ['SortModels', 'SortElements', 'SortAttributes']
available_columns = [col for col in columns_to_check if col in all_df.columns]
# 
# Check if the necessary sort columns are available
columns_to_check = ['SortModels', 'SortElements', 'SortAttributes']
available_columns = [col for col in columns_to_check if col in all_df.columns]

# Clean and prepare the data for sorting
for col in available_columns:
    # Convert to string only if the column is object (string) type
    if all_df[col].dtype == 'object':
        all_df[col] = all_df[col].astype(str).str.strip().str.lower()
    # For numeric columns, fill NaN with a very large number to sort them last
    elif all_df[col].dtype.kind in 'iuf':  # integer, unsigned integer, or float
        all_df[col] = all_df[col].fillna(float('inf'))
    # For datetime columns, fill NaN with the maximum datetime
    elif all_df[col].dtype.kind == 'M':  # datetime
        all_df[col] = all_df[col].fillna(pd.Timestamp.max)
    # For boolean columns, fill NaN with False to sort them last
    elif all_df[col].dtype.kind == 'b':  # boolean
        all_df[col] = all_df[col].fillna(False)

# Now sort the DataFrame using the available sort columns
sorted_df_combined = all_df.sort_values(by=available_columns, ascending=[True] * len(available_columns))
sorted_df_combined = sorted_df_combined.drop_duplicates().reset_index(drop=True)
```

```python
def export_with_custom_widths(all_df_export, column_widths):
    if 'DateiName' in all_df_export.columns:
        unique_filenames = all_df_export['DateiName'].dropna().unique()

        with pd.ExcelWriter(f'Elementplan_BBL_{version}.xlsx', engine='xlsxwriter') as writer:
            workbook = writer.book

            default_format = workbook.add_format({
                'text_wrap': True,
                'valign': 'top',
            })
            default_column_width = 20

            grey_text_format = workbook.add_format({'font_color': '#d3d3d3'})

            black_text_format = workbook.add_format({
                'top': 1,
                'text_wrap': True,
                'valign': 'top',
                'font_color': 'black'
            })

            centered_format = workbook.add_format({
                'text_wrap': True,
                'valign': 'top',
                'align': 'center',
            })

            for filename in unique_filenames:
                filtered_df = all_df_export[all_df_export['DateiName'] == filename]
                print(filename)
                filtered_df['Sort'] = range(1, len(filtered_df) + 1)

                filtered_df.to_excel(writer, sheet_name=filename[:31], index=False, startrow=0)  # Start writing data from row 1

                worksheet = writer.sheets[filename[:31]]

                                # Add the new rows at the top
                #worksheet.write('A1', 'Elementplan BBL')
                #worksheet.write('A2', 'Version:')
                #worksheet.write('B2', version)
                #worksheet.write('A3', 'Datum:')
                #worksheet.write('B3', datetime.now().strftime('%Y-%m-%d'))  # Current date
                #worksheet.write('A4', "")
                # Row 3 is left empty


                for col_num, col_name in enumerate(filtered_df.columns):
                    width = column_widths.get(col_name, default_column_width)
                    
                    if 11 <= col_num <= 22 or col_name == 'Sort':
                        worksheet.set_column(col_num, col_num, width, centered_format)
                    else:
                        worksheet.set_column(col_num, col_num, width, default_format)

                worksheet.autofilter(0, 0, len(filtered_df), len(filtered_df.columns) - 1)

                # Freeze panes at C2
                worksheet.freeze_panes(1, 3)

                # Apply a thin top border and black text starting from column C if column C is different from above
                worksheet.conditional_format(1, 2, len(filtered_df), len(filtered_df.columns) - 1, {
                    'type': 'formula',
                    'criteria': '=$C2<>$C1',
                    'format': black_text_format
                })

                # Apply a line across all cells if column A is different from above
                worksheet.conditional_format(1, 0, len(filtered_df), len(filtered_df.columns) - 1, {
                    'type': 'formula',
                    'criteria': '=$A2<>$A1',
                    'format': black_text_format
                })

                # Apply a line from column F onward, even if the cells are empty
                worksheet.conditional_format(1, 5, len(filtered_df), len(filtered_df.columns) - 1, {
                    'type': 'formula',
                    'criteria': '=$F2<>$F1',
                    'format': black_text_format
                })

                # Always apply a thin line starting from column G onward, including empty cells
                worksheet.conditional_format(1, 6, len(filtered_df), len(filtered_df.columns) - 1, {
                    'type': 'no_errors',
                    'format': black_text_format
                })

                # Apply grey text formatting to columns A to F where the value is the same as the row above
                for col in range(6):
                    col_letter = chr(65 + col)
                    worksheet.conditional_format(1, col, len(filtered_df), col, {
                        'type': 'formula',
                        'criteria': f'=${col_letter}2=${col_letter}1',
                        'format': grey_text_format
                    })

        output_file_by_filename = f'Elementplan_BBL_{version}.xlsx'
    else:
        output_file_by_filename = "FileName column not found in the DataFrame"

    return output_file_by_filename

# Assuming `all_df_export` is available and needs to be passed along with the template
output_file_path = export_with_custom_widths(all_df_export, column_widths)
output_file_path
```
